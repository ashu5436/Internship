```python
#Using Chrome Automation
driver=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32 (1)\chromedriver.exe')
driver.get('https://www.amazon.in/') ##taking amazon website
```

```python
#Extracting reviews,title and ratings for the headsets
srch_items = ['boat rockerz']

title = []

review_text = []

ratings = []


for i in srch_items:

    srchbar = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")

    srchbar.send_keys(i)

    driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input").click() 

    time.sleep(1)

    item_url = ['https://www.amazon.in/boAt-Rockerz-255-Pro-Earphones/dp/B08TV2P1N8/ref=sr_1_2?crid=2F4TJRPLYFS4O&keywords=boat+rockerz+255+pro+plus&qid=1669960831&qu=eyJxc2MiOiIzLjI3IiwicXNhIjoiMi44NSIsInFzcCI6IjIuNzQifQ%3D%3D&sprefix=boat+rockerz+%2Caps%2C1375&sr=8-2']

    for i in driver.find_elements("xpath","//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):

        item_url.append(i.get_attribute('href'))

    for i in item_url:

        driver.get(i)

        time.sleep(1)

        for _ in range(2):

            driver.execute_script("window.scrollBy(0,1000)")

            time.sleep(1)

        # Click to see all the reviews

        try:

            driver.find_element("xpath","//a[@class='a-link-emphasis a-text-bold']").click()

        except:

            pass

        

        url_ = []

        try:

            two_str = driver.find_element("xpath","//table[@id='histogramTable']/tbody/tr[4]/td[2]/a")

            url_.append(two_str.get_attribute('href'))

        except: pass

        try:

            three_str = driver.find_element("xpath","//table[@id='histogramTable']/tbody/tr[3]/td[2]/a")

            url_.append(three_str.get_attribute('href'))

        except: pass

        try:

            one_str = driver.find_element("xpath","//table[@id='histogramTable']/tbody/tr[5]/td[2]/a")

            url_.append(one_str.get_attribute('href'))

        except: pass

        try:

            fiv_str = driver.find_element("xpath","//table[@id='histogramTable']/tbody/tr[1]/td[2]/a")

            url_.append(fiv_str.get_attribute('href'))

        except: pass

        try:

            four_str = driver.find_element("xpath","//table[@id='histogramTable']/tbody/tr[2]/td[2]/a")

            url_.append(four_str.get_attribute('href'))

        except: pass

        for i in url_:

            driver.get(i)

            time.sleep(1)

            for i in range(150):

                ids = driver.find_elements("xpath","//div[@class='a-section a-spacing-none review-views celwidget']/div")

                comment_ids = []

                for i in ids:

                    comment_ids.append(i.get_attribute('id'))



                for x in comment_ids:

                    # Extracting user ids from each user on a page

                    try:

                        review_title = driver.find_element("xpath",'//*[@id="' + x +'"]/div/div/div[2]/a[2]/span[1]')

                        title.append(review_title.get_attribute("innerHTML"))

                    except NoSuchElementException:

                        title.append('')



                    try:

                        rating = driver.find_element("xpath",'//*[@id="' + x +'"]/div/div/div[2]/a[1]/i/span[1]')

                        ratings.append(rating.get_attribute("innerHTML").split(' ')[0])

                    except NoSuchElementException:

                        ratings.append('')



                    try:

                        text = driver.find_element("xpath",'//*[@id="' + x +'"]/div/div/div[4]/span/span')

                        review_text.append(text.get_attribute("innerHTML"))

                    except NoSuchElementException:

                        review_text.append('')

                try:

                    driver.find_element("xpath","//ul[@class='a-pagination']/li[2]/a").click()

                    time.sleep(3)

                except : break
```
