```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
#I'm doing used car scraping from 'CARTRADE' website
driver=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32 (1)\chromedriver.exe')
driver.get('https://www.cartrade.com/buy-used-cars/#sc=-1&so=-1&pn=1')
```

```python
driver.maximize_window()
```

```python
#taking out first 45 pages and trying to itterate the same process for each pages
name1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//h2[@class="h2heading truncate"]')
    for i in a:
        name1.append(i.text)
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
name1
```

```python
#Removing the year from the above attribute
Name1=[s[5:] for s in name1]
Name1
```

```python
Brand1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//h2[@class="h2heading truncate"]')
    for i in a:
        Brand1.append(i.text.split()[1])
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
Brand1
```

```python
len(Brand1)
```

```python
manufacturing_year1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//h2[@class="h2heading truncate"]')
    for i in a:
        manufacturing_year1.append(i.text.split()[0])
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
manufacturing_year1
```

```python
Price1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//div[@class="cr_prc"]')
    for i in a:
        Price1.append(i.text.split()[0])
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
Price1
```

```python
len(Price1)
```

```python
driven_kilometers1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//div[@class="info_cr_new"]')
    for i in a:
        driven_kilometers1.append(i.text.split('\n')[0])
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
driven_kilometers1
```

```python
Fuel1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//div[@class="info_cr_new"]')
    for i in a:
        Fuel1.append(i.text.split('\n')[2])
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
Fuel1
```

```python
Location1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//div[@class="info_cr_new"]')
    for i in a:
        Location1.append(i.text.split('\n')[4])
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
Location1
```

```python
product_urls1=[]
for page in range(0,45):
    a=driver.find_elements(By.XPATH,'//a[@style="font-weight: bold;"]')
    for i in a:
        product_urls1.append(i.get_attribute('href'))
    next_page=driver.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[2]/div[4]/div[2]/div/div/ul/li[4]')
    next_page.click()
    time.sleep(1)
product_urls1
```

```python
len(product_urls1)
```

```python
#here i've divided the scraping of owner attribute in 2 portion as my interent connectivity is not that good to scrap down the whole data from each link in one go
No_of_Owner1 = []   
for i in product_urls1[0:540]:
    driver.get(i)
    time.sleep(1)
    try:
        ow=driver.find_element(By.XPATH,'/html/body/div[2]/div[8]/div[1]/div[2]/div[1]/div[4]/table')
        No_of_Owner1.append(ow.text.split('\n')[4].split()[3])
    except NoSuchElementException:
        No_of_Owner1.append('-')
```

```python
No_of_Owner1
```

```python
len(No_of_Owner1)
```

```python
No_of_Owner2 = []   
for i in product_urls1[540:]:
    driver.get(i)
    time.sleep(1)
    try:
        ow=driver.find_element(By.XPATH,'/html/body/div[2]/div[8]/div[1]/div[2]/div[1]/div[4]/table')
        No_of_Owner2.append(ow.text.split('\n')[4].split()[3])
    except NoSuchElementException:
        No_of_Owner2.append('-')
```

```python
#final column for Owner
No_of_Owner2
```

```python
len(No_of_Owner2)
```

```python
No_of_Owner=No_of_Owner1+No_of_Owner2
No_of_Owner
```
