```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
#I'm using Quikr website for scraping used car data
driver=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32 (1)\chromedriver.exe')
driver.get('https://www.quikr.com/cars/used+cars+all-india+z1399')
```

```python
driver.maximize_window()
```

```python
driver.execute_script('window.scrollBy(0,1500)')
```

```python
#Scraping the name of the car's
name=[]
for i in driver.find_elements(By.XPATH,'//div[@class="car-info__footer"]'):
    name.append(i.text.split('\n')[0].split('-')[0])
name
```

```python
#Scraping the manufacturing_year of the car's
manufacturing_year=[]
for i in driver.find_elements(By.XPATH,'//div[@class="car-info__footer"]'):
    manufacturing_year.append(i.text.split('\n')[0].split('-')[1])
manufacturing_year
```

```python
len(manufacturing_year)
```

```python
#Scraping kilometers driven of each car
driven_kilometers=[]
for i in driver.find_elements(By.XPATH,'//div[@class="prime-features"]'):
    driven_kilometers.append(i.text.split('/')[0])
driven_kilometers
```

```python
len(driven_kilometers)
```

```python
#Scraping fuel type of each car
Fuel=[]
for i in driver.find_elements(By.XPATH,'//div[@class="prime-features"]'):
    Fuel.append(i.text.split('/')[1])
Fuel
```

```python
len(Fuel)
```

```python
#Scraping Price of each car
Price=[]
for i in driver.find_elements(By.XPATH,'//div[@class="price"]'):
    Price.append(i.text)
Price
```

```python
len(Price)
```

```python
#Scraping Locations of each car
Location=[]
for i in driver.find_elements(By.XPATH,'//footer[@class="qc-ads__card--footer"]'):
    Location.append(i.text.split('\n')[1])
Location
```

```python
len(Location)
```

```python
#scraping the url's of each car as some of the information are hidden inside link
product_urls = []
url=driver.find_elements(By.XPATH,'//a[@class="qc-ads__card"]')   
for i in url:
    product_urls.append(i.get_attribute('href'))                  
product_urls
```

```python
len(product_urls)
```

```python
#scraping the data of owners in several parts as the data is large
No_of_Owner1 = []   
for i in product_urls[0:500]:
    driver.get(i)
    time.sleep(1)
    try:
        view=driver.find_element(By.XPATH,'/html/body/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/span')
        view.click()
        ow=driver.find_element(By.XPATH,'//ul[@class="sc-1sbb775-0 lhcfIZ dF fwW jcSB"]')
        No_of_Owner1.append(ow.text.split('\n')[-1])
    except NoSuchElementException:
        No_of_Owner1.append('-')
```

```python
No_of_Owner2 = []   
for i in product_urls[500:1000]:
    driver.get(i)
    time.sleep(1)
    try:
        view=driver.find_element(By.XPATH,'/html/body/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/span')
        view.click()
        ow=driver.find_element(By.XPATH,'//ul[@class="sc-1sbb775-0 lhcfIZ dF fwW jcSB"]')
        No_of_Owner2.append(ow.text.split('\n')[-1])
    except NoSuchElementException:
        No_of_Owner2.append('-')
```

```python
#final column for total owners
No_of_Owner=No_of_Owner1+No_of_Owner2+No_of_Owner3+No_of_Owner4
No_of_Owner
```
