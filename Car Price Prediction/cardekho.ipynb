```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
#scraping rest of the data from 'CARDEKHO' website
driver=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32 (1)\chromedriver.exe')
driver.get('https://www.cardekho.com/used-cars+in+ahmedabad')
```

```python
Brand=[]
for i in driver.find_elements(By.XPATH,'//h3[@class="title"]'):
    Brand.append(i.text.split()[1])
Brand
```

```python
name=[]
for i in driver.find_elements(By.XPATH,'//h3[@class="title"]'):
    name.append(i.text)
name
```

```python
name=[s[5:] for s in name]
name
```

```python
len(name)
```

```python
manufacturing_year=[]
for i in driver.find_elements(By.XPATH,'//h3[@class="title"]'):
    manufacturing_year.append(i.text.split()[0])
manufacturing_year
```

```python
driven_kilometers1=[]
for i in driver.find_elements(By.XPATH,'//div[@class="dotsDetails"]'):
    driven_kilometers1.append(i.text.split()[0])
driven_kilometers1
```

```python
Fuel1=[]
for i in driver.find_elements(By.XPATH,'//div[@class="dotsDetails"]'):
    Fuel1.append(i.text.split()[3])
Fuel1
```

```python
Price1=[]
for i in driver.find_elements(By.XPATH,'//div[@class="Price hover"]'):
    Price1.append(i.text)
Price1
```

```python
product_urls1 = []
url=driver.find_elements(By.XPATH,'//h3[@class="title"]//a')   
for i in url:
    product_urls1.append(i.get_attribute('href'))                  
product_urls1
```

```python
Location1 = []   
for i in product_urls1:
    driver.get(i)
    time.sleep(1)
    try:
        ow=driver.find_element(By.XPATH,'//span[@class="value-text truncate-value hover posR"]')
        Location1.append(ow.text.split()[0])
    except NoSuchElementException:
        Location1.append('-')
Location1
```

```python
No_of_Owner1 = []   
for i in product_urls1:
    driver.get(i)
    time.sleep(1)
    try:
        ow=driver.find_element(By.XPATH,'//ul[@class="gsc_row detailsList"]')
        No_of_Owner1.append(ow.text.split('\n')[13])
    except NoSuchElementException:
        No_of_Owner1.append('-')
No_of_Owner1        
```

```python
print(len(Brand),len(name),len(manufacturing_year),len(driven_kilometers1),len(Fuel1),len(Price1),len(Location1),len(No_of_Owner1),len(product_urls1))
```

```python
import pandas as pd
df=pd.DataFrame({'Brand':Brand,'Car Name':name,'Manufacturing Year':manufacturing_year,'Driven KMs':driven_kilometers1,'Fuel Type':Fuel1,'Car Price':Price1,'Location':Location1,'Total Owners':No_of_Owner1,'URL"s':product_urls1})
df
```

```python
df.to_excel('Cardekho_used_car.xlsx')
```
