```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
#finding out the used car informations from 'CARWALE' website
driver=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32 (1)\chromedriver.exe')
driver.get('https://www.carwale.com/used/cars-in-delhi/#city=10&pc=10&sc=-1&so=-1&pn=1')
```

###### in this notebook i'm scraping the used car data from 'CARWALE' website and i will findout each attributes according to different locations in India. Here i've used 8 different location for the same.

##### 1

```python
name1=[]
for i in driver.find_elements(By.XPATH,'//span[@class="spancarname card-detail-block__title-text-with-year"]'):
    name1.append(i.text)
name1
```

```python
len(name1)
```

```python
manufacturing_year1=[]
for i in driver.find_elements(By.XPATH,'//span[@class="slYear card-detail-block__title-text-with-year"]'):
    manufacturing_year1.append(i.text)
manufacturing_year1
```

```python
driven_kilometers1=[]
for i in driver.find_elements(By.XPATH,'//span[@class="slkms"]'):
    driven_kilometers1.append(i.text)
driven_kilometers1
```

```python
Fuel1=[]
for i in driver.find_elements(By.XPATH,'//span[@class="fuel"]'):
    Fuel1.append(i.text)
Fuel1
```

```python
Price1=[]
for i in driver.find_elements(By.XPATH,'//span[@class="rupee-lac slprice"]'):
    Price1.append(i.text)
Price1
```

```python
Location1=[]
for i in driver.find_elements(By.XPATH,'//span[@class="cityName"]'):
    Location1.append(i.text)
Location1
```

```python
product_urls1 = []
url=driver.find_elements(By.XPATH,'//a[@id="linkToDetails"]')   
for i in url:
    product_urls1.append(i.get_attribute('href'))                  
product_urls1
```

```python
Brand1 = []   
for i in product_urls1:
    driver.get(i)
    time.sleep(1)
    try:
        br=driver.find_element(By.XPATH,'//div[@class="breadcrumb margin-bottom15"]')
        Brand1.append(br.text.split()[3])
    except NoSuchElementException:
        Brand1.append('-')
```

```python
Brand1
```

```python
No_of_Owner1 = []   
for i in product_urls1:
    driver.get(i)
    time.sleep(1)
    try:
        ow=driver.find_element(By.XPATH,'//*[@id="overview"]/div/ul/li[9]/div[2]')
        No_of_Owner1.append(ow.text)
    except NoSuchElementException:
        No_of_Owner1.append('-')
```

```python
No_of_Owner1
```

```python
driver.close()
```
